```python
import json
from tf.app import use
A = use('etcbc/bhsa')
F = A.TF.api.F
L = A.TF.api.L

RESULTS_PER_PAGE = 10
current_page = 0

# Load data
with open("data.json", 'r') as f:
    data = json.load(f)
# Convert JSON keys to integers.
data = { int(k): v for k, v in data.items() }

# make list of imperative forms
nodes = A.search('''
book book=Reges_I|Reges_II
    chapter
        clause
            word vt=impv
''')

print(f"{len(nodes)} occurrences of imperative")

def go_to_page(page_num):
    global current_page
    current_page = page_num

    start = current_page * RESULTS_PER_PAGE
    end = min((current_page + 1) * RESULTS_PER_PAGE, len(nodes))

    if start >= len(nodes):
        print("*** End of list ***")
    else:
        print("Showing results %d to %d (of 0-%d)"%(start, end - 1, len(nodes) - 1))
        A.table(nodes[start:end])

def get_node_id(item_num):
    # The list numbers displayed by `A.table()` are 1-based, not 0-based
    # so subtract 1
    return nodes[current_page * RESULTS_PER_PAGE + item_num - 1][3]

def tag(item_num, *tags):
    node_id = get_node_id(item_num)
    if data.get(node_id) == None:
        data[node_id] = []
    
    data[node_id].extend(tags)
```

```python
# the following sets of tags are intended to improve differentiation and capture more information
# ===============================================================================================

# whose wish/desire is relevant?
SPK_WISH = "expresses wish of speaker"
SUBJ_WISH = "expresses or assumes wish of subject (who is not speaker)"
ADDR_WISH = "expresses or assumes wish of addressee (who is not subject)"
MORAL = "expresses moral obligation rather than will"

# agency refers to who is in control or who is exerting their will through what is said
SPK_AGENCY = "expresses agency of speaker"
ADDR_AGENCY = "expresses/appeals to agency of addressee"
NO_AGENCY = "does not express/appeal to agency of speaker or addressee"

# authority here refers to the social setting regardless of speech content
HIGHER = "speaker has higher authority than addressee"
LOWER = "speaker has lower authority than addressee"
EQUAL = "speaker has roughly equal authority as addressee"
A_UNCLEAR = "authority is unclear or ambivalent"

DEONTIC = "certainly or probably deontic"
D_QUESTIONABLE = "questionable whether verb expresses deontic modality"

# TODO: remove new tags to prevent old data currupting new data when changes are made to this notebook
```

# Summary

This section summarizes the tags used to mark deontic modality in this notebook, specifically the combinations in which they occur.

```python
# compile list
results = A.search('''
book book=Reges_I|Reges_II
    chapter
        clause
            word vt=impv
''')

WISH_AGENCY = {
    SPK_WISH, SUBJ_WISH, ADDR_WISH, MORAL,
    SPK_AGENCY, ADDR_AGENCY, NO_AGENCY}

AUTHORITY = {HIGHER, LOWER, EQUAL, A_UNCLEAR}

# dict
# - key:   list of deontic tags
# - value: list of verbs with these deontic tags
deontic_categories = {}

for bk, ch , cl, word in results:
    # extract word's deontic tags for "wish" and "agency" to new list
    deontic_tags = [tag for tag in data[word] if tag in WISH_AGENCY]
    # convert to tuple for hashing
    deontic_tags = tuple(deontic_tags)

    if len(deontic_tags) > 0:
        # store deontic verb in dict
        if deontic_tags not in deontic_categories:
            deontic_categories[deontic_tags] = []

        deontic_categories[deontic_tags].append(word)

# print existing tag combinations, sorted from more common to less common
by_len = sorted(deontic_categories.items(), key=lambda item: -len(item[1]))
for tags, verbs in by_len:
    print(tags, len(verbs))
```

The following function summarizes the stats for a given category and displays the first 5 examples.

```python
def summarize_category(*tags):
    tags = tuple(tags)
    # dict (key: authority tag; value: list of verb nodes) 
    by_auth = {auth: [] for auth in AUTHORITY}

    count = 0
    for word in deontic_categories[tags]:
        # extract word's tag for authority
        auth = [tag for tag in data[word] if tag in AUTHORITY][0]
        # store word
        by_auth[auth].append(word)
        count += 1
    
    print(tags)
    print(f"{count} occurrences total")
    for auth, words in by_auth.items():
        print(f"{len(words):3}x {auth}")

    verbs_with_clauses = [
        (L.u(node, otype="clause")[0], node)
        for node in deontic_categories[tags]]
    A.table(verbs_with_clauses[:5])
```
